## **6. Plotting**

````python
x  = np.arange(0, 10, 0.01)
y1 = np.sin(x)
y2 = np.cos(x) * np.exp(-x * 0.3)

plt.plot(x, y1)
plt.plot(x, y2)
plt.grid()
plt.show()
````

````python
plt.subplot(211)
plt.plot(x, y1)
plt.grid()

plt.subplot(212)
plt.plot(x, y2)
plt.grid()
plt.show()
````
