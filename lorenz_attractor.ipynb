# About this code
So, this code, as you can see, is initated to solve the differential equations. I choose Lorenz attractor bacause it looks cool and it's a stiff system. I want to know, what are the problem that will ocure if we use a certain integration method. 

In this code, I use scipy.integrate.solve_ivp to solve the problem. It has method argument, so that user can choose what integration method they want to use. The default method is RK45 method.

# Lorenz attractor
You can read this https://en.wikipedia.org/wiki/Lorenz_system

Or, briefly, in this code we will solve this equation simultaniously

\begin{align}
        \frac{dx}{dt}&=\sigma (y-x)\\
        \frac{dy}{dt}&=x(\rho - z)-y\\
        \frac{dz}{dt}&=xz-\beta z
    \end{align}

With initial value, (x, y, z) = (1, 2, 3) and parameter ($\sigma, \rho, \beta$) = (10, 28, 8/3).


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from mpl_toolkits import mplot3d

def dSdt(t, S, sigma, rho, beta):
    x, y, z = S
    return [sigma*(y - x),
            x*(rho-z)-y,
            x*y-beta*z]

fig = plt.figure(dpi=100)
max_t = 20    # maksimum time sequence 
lw = 1        # line width

# first plot
sol = solve_ivp(dSdt, t_span=(0, max_t), y0=[1,2,3], method="RK45", args=(10,28,8/3))
x, y, z = sol.y[0], sol.y[1], sol.y[2]
ax1 = fig.add_subplot(3, 2, 1, projection="3d")
ax1.set_title("Integration method: RK45")
ax1.set_axis_off()
ax1.plot3D(x, y, z, lw=lw)

# second plot
sol = solve_ivp(dSdt, t_span=(0, max_t), y0=[1,2,3], method="RK23", args=(10,28,8/3))
x, y, z = sol.y[0], sol.y[1], sol.y[2]
ax2 = fig.add_subplot(3, 2, 2, projection="3d")
ax2.set_title("Integration method: RK23")
ax2.plot3D(x, y, z, lw=lw)
ax2.set_axis_off()

# third plot
sol = solve_ivp(dSdt, t_span=(0, max_t), y0=[1,2,3], method="DOP853", args=(10,28,8/3))
x, y, z = sol.y[0], sol.y[1], sol.y[2]
ax3 = fig.add_subplot(3, 2, 3, projection="3d")
ax3.set_title("Integration method: DOP853")
ax3.plot3D(x, y, z, lw=lw)
ax3.set_axis_off()

# fouth plot
sol = solve_ivp(dSdt, t_span=(0, max_t), y0=[1,2,3], method="Radau", args=(10,28,8/3))
x, y, z = sol.y[0], sol.y[1], sol.y[2]
ax4 = fig.add_subplot(3, 2, 4, projection="3d")
ax4.set_title("Integration method: Radau")
ax4.plot3D(x, y, z, lw=lw)
ax4.set_axis_off()

# fifth plot
sol = solve_ivp(dSdt, t_span=(0, max_t), y0=[1,2,3], method="BDF", args=(10,28,8/3))
x, y, z = sol.y[0], sol.y[1], sol.y[2]
ax5 = fig.add_subplot(3, 2, 5, projection="3d")
ax5.set_title("Integration method: BDF")
ax5.plot3D(x, y, z, lw=lw)
ax5.set_axis_off()

# sixth plot
sol = solve_ivp(dSdt, t_span=(0, max_t), y0=[1,2,3], method="LSODA", args=(10,28,8/3))
x, y, z = sol.y[0], sol.y[1], sol.y[2]
ax6 = fig.add_subplot(3, 2, 6, projection="3d")
ax6.set_title("Integration method: LSODA")
ax6.plot3D(x, y, z, lw=lw)
ax6.set_axis_off()

plt.tight_layout()
plt.show()
```
