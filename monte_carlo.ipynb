**1. Menetukan Luasan Lingkaran dengan Loop**

```python
import numpy as np
import matplotlib.pyplot as plt

dn = 0.1                # banyak titik yang digunakan = (1/dn)^2
x = np.arange(0,1,dn)
y = x
num = 0                 # banyak titik yang memenuhi syarat
den = len(x)**2
vval = []

for i in x:
  for j in y:
    if i**2 + j**2 < 1:
      num += 1
      val = num/den * 4
      vval.append(val)
    
plt.plot(vval, label="$\pi$ estimation")
plt.hlines(np.pi, 0, len(vval), label="$\pi$ value")
plt.legend()
plt.grid()
plt.show()
```

**2. Menentukan Luasan Lingkaran dengan Simulasi Monte Carlo**

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100   # banyak titik yang digunakan
num = 0   # banyak titik yang memenuhi syarat
vval = []

for i in range(n):
  x = np.random.uniform(-1, 1)
  y = np.random.uniform(-1, 1)
  if x**2 + y**2 < 1:
    num += 1
    den = i
    if i == 0:
      val = 0
    else:
      val = num/den * 4
    vval.append(val)
    
plt.plot(vval, label="$\pi$ estimation")
plt.hlines(np.pi, 0, len(vval), label="$\pi$ value")
plt.legend()
plt.grid()
plt.show()
```

**3. Kesimpulan**

Metode Monte Carlo menghampiri nilai $\pi$ lebih cepat daripada metode loop dengan banyak titik tinjauan yang sama.
