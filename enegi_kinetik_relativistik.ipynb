```python
# memberikan keluaran berupa energi kinetik partikel bermassa 1 kg
# ditinjau dari mekanika klasik dan mekanika relativistik

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 1, 0.01)
ek_rel = 1/(np.sqrt(1 - x**2)) - 1
ek_cla = 1/2 * x**2
x_axis = ["0", "0.2c", "0.4c", "0.6c", "0.8c", "c"] 

plt.figure(dpi=150)
plt.xticks(np.arange(0, 1.2, 0.2), x_axis)

plt.plot(x, ek_rel, label="Relativistik")
plt.plot(x, ek_cla, label="Klasik")

plt.xlabel("Kelajuan (m/s)")
plt.ylabel("Energi Kinetik (J)")

plt.legend()
plt.grid()
plt.show()
```
