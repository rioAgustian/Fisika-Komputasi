## ***Fine Tunning* Grafik**

*Fine tunning* adalah memoles grafik agar lebih mudah dipahami atau dianalisis. Agar lebih mudah dipahami, kita bisa menambahkan keterangan-keterangan di grafik kita. Misalnya, kita dapat memberi judul grafik, menamai axis, membuat grid, memberi nama fungsi, dan lain sebagainya.

```python
x = np.arange(0, 10, 1)
y = 2 * x

plt.figure(dpi=300, figsize=(8, 8))
plt.plot(x, y, label="V = RI")

plt.title("Grafik hubungan tegangan dangan arus listrik")
plt.xlabel("I (ampere)")
plt.ylabel("V (volt)")

plt.legend()
plt.grid()
plt.show()
```

## **Membuat Noise**

Dalam melakukan praktikum, pastinya ada noise atau gangguan yang membuat data yang didapat tidak pas dengan model yang dimiliki. Kita dapat membuat noise di Python dengan menggunakan `numpy.random`

```python
from numpy.random import randn
deviasi = 0.5
noise = deviasi * randn(1, 1000)

x = np.arange(0, 10, 0.01)
y = 2 * x + noise

plt.scatter(x, y, marker="x")
plt.grid();
```
