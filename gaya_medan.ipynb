```python
import matplotlib.pyplot as plt

# parameter (dalam SI)
dt = 3.33e-10
t = 0
t_max = 3.33e-8
v = 3e6
e = 1.6e-19
E = 200
m = 9.11e-31
a = e * E / m

# list kosong
xx = []
yy = []
tt = []

print("t \t x \t y")
while t <= t_max:
  y = 1/2 * a * t**2
  x = v * t
  t = t + dt

  # daftar t, x dan y
  print(t, x, y)

  # mengisi list
  xx.append(x)
  yy.append(y)
  tt.append(t)

# plot grafik
plt.plot(xx, yy, label="Lintasan elektron")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```
